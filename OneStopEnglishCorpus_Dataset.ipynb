```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import os
```

#### ELEM

```python
url = "https://github.com/nishkalavallabhi/OneStopEnglishCorpus/tree/master/Texts-SeparatedByReadingLevel/Ele-Txt"
```

```python
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, "html.parser")
```

```python
soup
```

```python
links = soup.find_all("a")
data_links = [link["href"] for link in links if ".txt" in link["href"]]
data_links
```

```python
for link in data_links:
    full_url = f"https://raw.githubusercontent.com/{link.replace('/blob', '')}"
    #print(full_url)
    file_name = os.path.splitext(link.split("/")[-1])[0]
    response = requests.get(full_url)
    text_content = response.text
    with open(file_name, "w") as f:
        f.write(text_content)
```

```python
df_elem = pd.DataFrame(columns=["url", "text"])
```

```python
for link in data_links:
    file_name = os.path.splitext(link.split("/")[-1])[0]
    with open(file_name, "r") as f:
        text_content = f.read()
    df_elem = df_elem.append({"url": link, "text": text_content}, ignore_index=True)
```

```python
df_elem
```

```python
df_elem["level"] = "Elementary"
```
