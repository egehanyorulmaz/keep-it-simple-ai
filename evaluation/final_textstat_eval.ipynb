```python
import pandas as pd
import textstat
```

```python
final_df = pd.read_csv('final_data.csv')
final_df
```

```python
final_df = final_df.dropna(axis=0)
```

```python
def calc_gfog(text):
    return textstat.gunning_fog(text)

def calc_flesch(text):
    return textstat.flesch_reading_ease(text)

def calc_dale(text):
    return textstat.dale_chall_readability_score(text)
```

```python
def textstat_eval(df):
    
    df['context_gfog'] = round(df['context'].apply(calc_gfog),2)
    df['response_gfog'] = round(df['response'].apply(calc_gfog),2)
    df['model_7b_chat_response_gfog'] = round(df['model_7b_chat_response'].apply(calc_gfog),2)
    df['oob_model_response_gfog'] = round(df['oob_model_response'].apply(calc_gfog),2)
    df['model_v2_response_gfog'] = round(df['model_v2_response'].apply(calc_gfog),2)

    df['context_fl'] = round(df['context'].apply(calc_flesch),2)
    df['response_fl'] = round(df['response'].apply(calc_flesch),2)
    df['model_7b_chat_response_fl'] = round(df['model_7b_chat_response'].apply(calc_flesch),2)
    df['oob_model_response_fl'] = round(df['oob_model_response'].apply(calc_flesch),2)
    df['model_v2_response_fl'] = round(df['model_v2_response'].apply(calc_flesch),2)

    df['context_dale'] = round(df['context'].apply(calc_dale),2)
    df['response_dale'] = round(df['response'].apply(calc_dale),2)
    df['model_7b_chat_response_dale'] = round(df['model_7b_chat_response'].apply(calc_dale),2)
    df['oob_model_response_dale'] = round(df['oob_model_response'].apply(calc_dale),2)
    df['model_v2_response_dale'] = round(df['model_v2_response'].apply(calc_dale),2)

    stats_gfog_mean = df.groupby('instruction')[['context_gfog','response_gfog', 'model_7b_chat_response_gfog', 'oob_model_response_gfog','model_v2_response_gfog']].mean().reset_index() 
    stats_fl_mean = df.groupby('instruction')[['context_fl','response_fl', 'model_7b_chat_response_fl','oob_model_response_fl','model_v2_response_fl']].mean().reset_index()
    stats_dale_mean = df.groupby('instruction')[['context_dale','response_dale', 'model_7b_chat_response_dale','oob_model_response_dale','model_v2_response_dale']].mean().reset_index()
     
    stats_gfog_std = df.groupby('instruction')[['context_gfog','response_gfog', 'model_7b_chat_response_gfog', 'oob_model_response_gfog','model_v2_response_gfog']].std().reset_index()
    stats_fl_std = df.groupby('instruction')[['context_fl','response_fl', 'model_7b_chat_response_fl','oob_model_response_fl','model_v2_response_fl']].std().reset_index()
    stats_dale_std = df.groupby('instruction')[['context_dale','response_dale', 'model_7b_chat_response_dale','oob_model_response_dale','model_v2_response_dale']].std().reset_index()
    
    return stats_gfog_mean,stats_fl_mean,stats_dale_mean,stats_gfog_std,stats_fl_std,stats_dale_std
```

```python
results = textstat_eval(df=final_df)
```

```python
stats_gfog_mean = results[0]
stats_fl_mean = results[1]
stats_dale_mean = results[2]
stats_gfog_std = results[3]
stats_fl_std = results[4]
stats_dale_std = results[5]
```
