```python
import requests
```

```python
url_A1 = 'https://github.com/AMontgomerie/CEFR-English-Level-Predictor/tree/main/data/A1'
response_A1 = requests.get(url_A1)
```

```python
raw_url_A1 = 'https://raw.githubusercontent.com/AMontgomerie/CEFR-English-Level-Predictor/main/data/A1'

import re
import pandas as pd

A1_df = pd.DataFrame(columns=["Texts"])
for file_name in response_A1.content.splitlines():
    file_name = file_name.decode('utf-8')
    
    matches = re.findall(r'href\s*=\s*([^\s]+\.txt)', file_name)
    if (len(matches) != 0):
        name = matches[0]        
        name = name.split(">")[1]
        
        file_url = raw_url_A1 + '/' + name
        response = requests.get(file_url)
        text = response.content.decode('utf-8')
        A1_df = A1_df.append({"Texts": text}, ignore_index = True)
```

```python
A1_df['CEFR_Level'] = "A1"
A1_df
```
