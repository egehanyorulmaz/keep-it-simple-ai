```python
import pandas as pd
```

```python
base_url = "https://www.newsinlevels.com/level/level-1/page/"
page_num = 1
article_links = []

while True:
    # construct url for current page
    url = base_url + str(page_num) + "/"
    
    # make request and parse page content
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    # find all article links on page and add them to the list
    articles = soup.find_all('a', href=True)
    for article in articles:
        if '/products/' in article['href']:
            article_links.append(article['href'])
    
    # check if there is a "Next" button on the page
    next_button = soup.find('a', text='›')
    if next_button:
        page_num += 1
    else:
        break

# create dataframe with article links
x = pd.DataFrame({'level_url': article_links})
```

```python
x['level'] = x['level_url'].str.extract(r'level-(\d)')
x
```

```python
def scrape_article_text(url):
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content, 'html.parser')
    article_text = soup.find('div', id='nContent').get_text().strip()
    return article_text
```

```python
def scrape_article_text2(url):
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content, 'html.parser')
    article_text = ''
    nContent = soup.find('div', id='nContent')
    if nContent:
        article_text = nContent.get_text().strip()
    return article_text
```

```python
from pandarallel import pandarallel

# Initialize pandarallel
pandarallel.initialize(progress_bar=False, nb_workers=3)
```

```python
x['text'] = x['level_url'].parallel_apply(scrape_article_text2)
```

```python
x['text'] = x['text'].str.split('Difficult words:').str[0] # remove text after "Difficult words:"
x['text'] = x['text'].str[17:] # remove date at beginning of text
x['text'] = x['text'].replace('\r\n', ' ').replace('\n', ' ') 
```
